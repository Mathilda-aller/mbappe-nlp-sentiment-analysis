```python
import pandas as pd
df_pg = pd.read_excel('output_psg.xlsx')
#print(df_pg.info())      

# 查看缺失值情况
#print(df_pg.isnull().sum())      

# 删除包含缺失值的行（如果评论内容缺失）
df_pg = df_pg.dropna(subset=['textDisplay'])
# 去除重复评论
df_pg = df_pg.drop_duplicates(subset=['textDisplay'])
# 去除无意义的评论，例如：只有表情符号、无实际内容的评论
import re

# 只保留包含字母、数字或有文字的评论
df_pg = df_pg[df_pg['textDisplay'].str.contains('[a-zA-Z0-9]', na=False)]

# 过滤掉表情符号（如果评论只包含表情符号或无意义字符）
df_pg = df_pg[~df_pg['textDisplay'].str.contains('[\U00010000-\U0010ffff]', na=False)]

#只保留法语评论
from langdetect import detect, LangDetectException
def safe_detect(text):
    try:
        if len(text) < 3:
            return 'und'  
        return detect(text)
    except LangDetectException:
        return 'und' 

df_pg['language'] = df_pg['textDisplay'].apply(lambda x: safe_detect(x))
df_pg = df_pg[df_pg['language'] == 'fr']

# 去除长度小于 5 个字符的评论
df_pg = df_pg[df_pg['textDisplay'].str.len() > 5]


# 正则表达式筛选，删除含有广告链接、推销词语等的评论
df_pg = df_pg[~df_pg['textDisplay'].str.contains('http|buy|discount|click', case=False, na=False)]
# 将评论文本转换为小写字母
df_pg['textDisplay'] = df_pg['textDisplay'].str.lower()
# 去除多余的空格
df_pg['textDisplay'] = df_pg['textDisplay'].str.strip()
# 检查文本中的特殊字符，修复或去除
df_pg['textDisplay'] = df_pg['textDisplay'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))

# 保存清理后的数据
df_pg.to_excel('cleaned_output_psg.xlsx', index=False)
# 查看清理后的数据
print(df_pg.head())
```
